```python
import sqlite3
from faker import Faker
from datetime import datetime
import random

# Inicializar Faker para generar datos aleatorios en español
fake = Faker('es_CO')

# Número de registros a generar
num_registros = 20000000

# Nombre de la base de datos SQLite
nombre_db = 'mi_base_de_datos.db'

def crear_tabla():
    """Crea la tabla en la base de datos si no existe."""
    conn = sqlite3.connect(nombre_db)
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS datos_personales (
            Identificacion TEXT PRIMARY KEY,
            Nombres TEXT,
            Apellidos TEXT,
            Fecha_Nacimiento TEXT,
            Edad INTEGER,
            Sexo TEXT,
            Estado_Civil TEXT,
            Correo_Electronico TEXT
        )
    ''')
    conn.commit()
    conn.close()

def generar_datos(num_registros):
    """Genera una lista de diccionarios con datos aleatorios."""
    datos = []
    for _ in range(num_registros):
        identificacion = fake.unique.random_number(digits=10)
        fecha_nacimiento = fake.date_of_birth(minimum_age=18, maximum_age=90).strftime('%Y-%m-%d')
        edad = datetime.now().year - datetime.strptime(fecha_nacimiento, '%Y-%m-%d').year
        sexo = random.choice(['Masculino', 'Femenino'])
        estado_civil = random.choice(['Soltero/a', 'Casado/a', 'Divorciado/a', 'Viudo/a', 'Unión Libre'])
        nombres = fake.first_name()
        apellidos = fake.last_name() + ' ' + fake.last_name()
        correo_electronico = fake.email()
        datos.append((
            str(identificacion),
            nombres,
            apellidos,
            fecha_nacimiento,
            edad,
            sexo,
            estado_civil,
            correo_electronico
        ))
        if _ % 10000 == 0:
            print(f"Generados {_} registros...")
    return datos

def insertar_datos(datos):
    """Inserta los datos en la base de datos."""
    conn = sqlite3.connect(nombre_db)
    cursor = conn.cursor()
    cursor.executemany('''
        INSERT INTO datos_personales (
            Identificacion, Nombres, Apellidos, Fecha_Nacimiento, Edad, Sexo, Estado_Civil, Correo_Electronico
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    ''', datos)
    conn.commit()
    conn.close()
    print("Todos los datos han sido insertados.")

if __name__ == "__main__":
    crear_tabla()
    print("Generando datos...")
    datos_generados = generar_datos(num_registros)
    print("Insertando datos en la base de datos...")
    insertar_datos(datos_generados)
    print(f"Base de datos '{nombre_db}' creada con {num_registros} registros.")
```
